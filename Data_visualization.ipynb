```python
flight_data = pd.read_csv('/content/drive/MyDrive/DataScience/flight_delays.csv',index_col = 'Month')
flight_data.head()
```

### BAR PLOT

```python
plt.figure(figsize = (10,8))
plt.title('Average Arrival Delay')
sns.barplot(x = flight_data.index , y = flight_data.WN)

# Adding labels
plt.xlabel('Months')
plt.ylabel('Average delay')
```

```python
plt.figure(figsize = (10,5))
plt.title('Average Arrival Delay')
sns.heatmap(data = flight_data , annot = True , cmap = 'BuGn')

# Adding labels
plt.xlabel('Flights')
plt.ylabel('Average delay')
```

#### CORELATION MATRIX

```python
flight_data.corr()
```

#### HEATMAP

```python
plt.figure(figsize = (10,5))
plt.title('Average Arrival Delay')
corr = flight_data.corr()
sns.heatmap(data = corr , annot = True , cmap = 'BuGn')

# Adding labels
plt.xlabel('Flights')
plt.ylabel('Average delay')
```

```python
insurance_data = pd.read_csv('/content/drive/MyDrive/DataScience/insurance.csv' )
insurance_data.head()
```

#### SCATTER PLOT

```python
plt.figure(figsize = (10,5))
plt.title('Insurance Data')

sns.scatterplot(x = insurance_data.age, y = insurance_data.charges , hue = insurance_data.smoker)
```

#### REGRESSION PLOT

```python
plt.figure(figsize = (10,5))
plt.title('Insurance Data')

sns.regplot(x = insurance_data.age, y = insurance_data.charges)
```

```python
plt.figure(figsize = (10,5))
sns.lmplot(data =insurance_data, x= 'age' , y= 'charges' , hue = 'smoker')
```
