```python
#["EVEN" if i%2==0 else "ODD" for i in range(10)]

# Concentrating on US and Non-US countries - List comprehension
x['native_country'] = ['other' if i!='United-States' else i for i in x.native_country]
x['native_country'].value_counts().sort_values(ascending = False)
```

```python
native_country_dummies = pd.get_dummies(x.native_country)
native_country_dummies.head()
```

```python
# Concentrating on US and Non-US countries - numpy.where()

x.native_country = np.where(x.native_country == 'United-States', 'United-States','other')
x.native_country.value_counts().sort_values(ascending = False)
```

```python
categorical_columns = [i for i in x.columns if x[i].dtype == 'object']
categorical_columns
```

```python
Transformed_data = pd.get_dummies(x[categorical_columns], dummy_na=False)
Transformed_data.head()
```

```python
x.drop(categorical_columns , axis =1 ,inplace=True)
```

```python
preprocessed_data = pd.concat([x,Transformed_data],axis=1)

preprocessed_data.head(3)

#null_counts = [(i,preprocessed_data[preprocessed_data[i].isnull()][i].size) for i in preprocessed_data.columns if preprocessed_data[preprocessed_data[i].isnull()][i].size > 0]

def null_check(df):
  null_counts = [(i,df[df[i].isnull()][i].size) for i in df.columns if df[df[i].isnull()][i].size > 0]
  return null_counts

null_check(preprocessed_data)

#[i for i in range(100) if (i%2==0 and i%5==0) ]
```

```python
null_check(preprocessed_data.dropna())
```
